```python
import pickle
import pandas as pd
from sklearn.preprocessing import RobustScaler
```

```python
df_diamond = pd.read_csv("data/test.csv")
df_diamond_sub = pd.read_csv("data/sample_submission.csv")
```

```python
df_diamond
```

```python
df_diamond = df_diamond.drop(["id","x","y","z","depth","table"], axis=1)
```

```python
df_diamond
```

```python
# Encoding Diamantes
with open("data/encodingclarity.pkl", 'rb') as diamond:
    encoding_clarity = pickle.load(diamond)
with open("data/encodingcolor.pkl", 'rb') as diamond:
    encoding_color = pickle.load(diamond)
with open("data/encodingcut.pkl", 'rb') as diamond:
    encoding_cut = pickle.load(diamond)
    
# Estandarización
with open('data/estandarizacion.pkl', 'rb') as estand:
    estandarizacion_diamond = pickle.load(estand)
```

```python
with open('data/mejor_modelo.pkl', 'rb') as modelo:
    modelo = pickle.load(modelo)
```

```python
df_diamond
```

```python
carat = pd.DataFrame(estandarizacion_diamond.transform(df_diamond[["carat"]]), 
                              columns = ["carat"])
carat
```

```python
df_diamond
```

```python
clarity_encoded = pd.DataFrame(encoding_clarity.transform(df_diamond[["clarity"]]), columns= ["clarity_encoded"])
color_encoded = pd.DataFrame(encoding_color.transform(df_diamond[["color"]]), columns= ["color_encoded"])
cut_encoded = pd.DataFrame(encoding_cut.transform(df_diamond[["cut"]]), columns= ["cut_encoded"])
```

```python
df_final = pd.concat([cut_encoded, color_encoded, clarity_encoded, carat], axis = 1)
df_final
```

```python
df_diamond_sub
```

```python
sample_submission = modelo.predict(df_final)
sample_submission
```

```python
sample_submission = pd.DataFrame(sample_submission)
sample_submission
```

```python
sample_submission = sample_submission.reset_index()
```

```python
sample_submission = sample_submission.rename(columns={"index":"id",0:"price"})
```

```python
sample_submission
```

```python
sample_submission.to_csv("data/sample_submission_model_1.csv",index=False)
```
